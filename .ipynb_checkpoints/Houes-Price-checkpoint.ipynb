```python
categorical_data = [col for col in df.columns if df[col].dtype == 'object']

from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
for column in categorical_data:
    df[column] = encoder.fit_transform(df[column].astype(str))
```

```python
train_df = df[:len(train)]
test_df = df[len(test) + 1:]
```

```python
len(train_df), len(train)
```

```python
len(test_df), len(test)
```

```python
X = train_df.drop(['SalePrice', 'Id'], axis=1)
y = train_df['SalePrice']

X_test = test_df.drop(['SalePrice', 'Id'], axis=1)
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_model = LinearRegression()
```

```python
linear_model.fit(X_train, y_train)
```

```python
pred = linear_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
MAE = mean_absolute_error(y_test, pred)
MSE = mean_squared_error(y_test, pred)
RMSE = np.sqrt(MSE)
R2 = r2_score(y_test, pred)

print(f"MAE: {MAE}")
print(f"MSE: {MSE}")
print(f"RMSE: {RMSE}")
print(f"R2: {R2}")
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
n_est = [50,100,150,200]

for n in n_est:
    rf = RandomForestRegressor(n_estimators=n, oob_score=True, random_state=101)
    rf.fit(X_train, y_train)
    pred = rf.predict(X_test)
    MAE = mean_absolute_error(y_test, pred)
    print(f"n_est: {n}, OOB: {rf.oob_score_}, MAE: {MAE}")
```

```python
rf_model = RandomForestRegressor(n_estimators=150, random_state=101)
```

```python
rf_model.fit(X_train, y_train)
```

```python
pred = rf_model.predict(X_test)
```

```python
MAE = mean_absolute_error(y_test, pred)
MSE = mean_squared_error(y_test, pred)
RMSE = np.sqrt(MSE)
R2 = r2_score(y_test, pred)

print(f"MAE: {MAE}")
print(f"MSE: {MSE}")
print(f"RMSE: {RMSE}")
print(f"R2: {R2}")
```

```python
from sklearn.svm import SVR
```

```python
svr_model = SVR(kernel='rbf')
```

```python
svr_model.fit(X_train, y_train)
```

```python
pred = svr_model.predict(X_test)
```

```python
MAE = mean_absolute_error(y_test, pred)
MSE = mean_squared_error(y_test, pred)
RMSE = np.sqrt(MSE)
R2 = r2_score(y_test, pred)

print(f"MAE: {MAE}")
print(f"MSE: {MSE}")
print(f"RMSE: {RMSE}")
print(f"R2: {R2}")
```

```python
from sklearn.model_selection import GridSearchCV
from xgboost import XGBRegressor
```

```python
xg = XGBRegressor(objective='reg:squarederror', random_state=101)
```

```python
param_grid = {
    'n_estimators': [50, 100, 150, 200],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0]
}
```

```python
grid_search = GridSearchCV(estimator=xg, param_grid=param_grid, cv=5, scoring='neg_mean_absolute_error', verbose=2, n_jobs=-1)
```

```python
grid_search.fit(X_train, y_train)
```

```python
best_params = grid_search.best_params_
print("Best parameters:", best_params)
```

```python
xg_model = XGBRegressor(**best_params, objective='reg:squarederror', random_state=101)
```

```python
xg_model.fit(X_train, y_train)
```

```python
pred = xg_model.predict(X_test)
```

```python
MAE = mean_absolute_error(y_test, pred)
MSE = mean_squared_error(y_test, pred)
RMSE = np.sqrt(MSE)
R2 = r2_score(y_test, pred)

print(f"MAE: {MAE}")
print(f"MSE: {MSE}")
print(f"RMSE: {RMSE}")
print(f"R2: {R2}")
```
